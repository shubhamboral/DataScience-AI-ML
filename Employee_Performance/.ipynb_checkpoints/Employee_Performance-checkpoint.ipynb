```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1.Load the dataset from a CSV file.

df = pd.read_csv(r"C:\Users\Bloodhound\Downloads\Github\DataScience-AI-ML\Employee_Performance\Employee_Performance.csv")
print(df.to_string()) 
```

```python
#2.Clean the data by handlling missing values formatting inconsistencies and removing duplicates.

df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df['Performance_Score'].fillna(df['Performance_Score'].median(), inplace=True)
print(df.to_string())
```

```python
#3. Extract insights by filtering employees based on experience, performance and salary trends

high_value_orders = df[df['Experience'] > 10]
print("\nExperienced Employees-:\n", high_value_orders)

high_value_orders1 = df[df['Performance_Score'] > 3]
print("\nPerformant Employees:\n", high_value_orders1)

high_value_orders2 = df[df['Salary'] > 100000]
print("\nHighly Paid Employees:\n", high_value_orders2)
```

```python
#4. Perform aggreagtion to analyze average salaries by department and experience levels.

avg=df['Salary'].mean()
print(avg)
```

```python
#5. Sort and filter top performers who should be considered for promotion and bonuses.

df_sorted1 = df.sort_values(by='Performance_Score', ascending=False)
print("\nDataFrame after sorting :\n", df_sorted1)
```

```python
#Save the New Cleaned Dataset

df.to_csv("Cleaned_Employee_Performance.csv",index=False)
```

```python
#Create key visualizations using mathplotlib and seaborn to analyze
# #salary distribution,performance score trends,experience vs salary corelation,department wise salary comparison

df = pd.read_csv("Cleaned_Employee_Performance.csv")

# Salary Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Salary'], kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# Performance Score Trends 
plt.figure(figsize=(8, 6))
sns.lineplot(x=df.index, y='Performance_Score', data=df) 
plt.title('Performance Score Trend')
plt.xlabel('Experience') 
plt.ylabel('Performance Score')
plt.show()

# Experience vs. Salary Correlation

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Experience', y='Salary', data=df)
plt.title('Experience vs. Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
correlation = df['Experience'].corr(df['Salary'])
print(f"Correlation between Experience and Salary: {correlation}")


# Department-wise Salary Comparison

plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Department-wise Salary Comparison')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=45, ha='right')  
plt.show()
```
