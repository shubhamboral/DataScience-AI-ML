```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Customer Purchase behaviour analysis

df = pd.read_csv(r"C:\Users\Bloodhound\Downloads\Github\DataScience-AI-ML\Customer_Purchases\Customer_Purchases.csv")
print(df.to_string())
```

```python
 #1.Clean the data (handle missing values and duplicates)

print(df.isnull().sum())

df['Total_Spending'].fillna(df['Total_Spending'].mean(), inplace=True)

df.drop_duplicates(inplace=True)

print("\nDataFrame after cleaning:\n", df)
```

```python
#2.Analyse spending patterns based on customer-age, gender and region.

# 1. Spending vs. Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Total_Spending', hue='Gender', data=df)
plt.title('Spending vs. Age')
plt.xlabel('Customer Age')
plt.ylabel('Total Spending')
plt.show()


# 2. Spending by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Total_Spending', data=df)
plt.title('Spending by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.show()


# 3. Spending by Region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Total_Spending', data=df, estimator=sum)
plt.title('Spending by Region')
plt.xlabel('Region')
plt.ylabel('Total Spending')
plt.xticks(rotation=45, ha='right')
plt.show()


# 4. Spending Distribution by Region and Gender
plt.figure(figsize=(12, 8))
sns.violinplot(x='Region', y='Total_Spending', hue='Gender', data=df, split=True)
plt.title('Spending Distribution by Region and Gender')
plt.xlabel('Region')
plt.ylabel('Total Spending')
plt.xticks(rotation=45, ha='right')
plt.show()

# Calculate average spending for each combination of age, gender, and region
print(df.groupby(['Age', 'Gender', 'Region'])['Total_Spending'].mean())
```

```python
#3. Find high-value customers (customers with high total spending)

customer_spending = df.groupby('Customer_ID')['Total_Spending'].sum()

threshold = customer_spending.quantile(0.9)

high_value_customers = customer_spending[customer_spending >= threshold]

print("High-value customers (top 10% spenders):\n", high_value_customers)
```

```python
#4. Identify seasonal trends in purchases.

# Convert 'Date' column to datetime objects
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

# Extract month and year information
df['Month'] = df['Purchase_Date'].dt.month
df['Year'] = df['Purchase_Date'].dt.year

# Group data by month and year, and sum total spending
seasonal_trends = df.groupby(['Year', 'Month'])['Total_Spending'].sum().reset_index()

# Create a line plot to visualize the seasonal trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Total_Spending', hue='Year', data=seasonal_trends, marker='o')
plt.title('Seasonal Trends in Purchases')
plt.xlabel('Month')
plt.ylabel('Total Spending')
plt.xticks(range(1, 13))  
plt.show()
```

```python
#5. Visualise key insights(bar charts, histograms,scatter plots)

# 1. Total Spending Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Total_Spending'], kde=True)
plt.title('Distribution of Total Spending')
plt.xlabel('Total Spending')
plt.ylabel('Frequency')
plt.show()


# 2. Average Spending by Age Group (using bar chart)
df['Age_Group'] = pd.cut(df['Age'], bins=range(0, 101, 10), right=False)
age_spending = df.groupby('Age_Group')['Total_Spending'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Age_Group', y='Total_Spending', data=age_spending)
plt.title('Average Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Total Spending')
plt.xticks(rotation=45, ha='right')
plt.show()


# 3. Relationship between Age and Total Spending (scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Total_Spending', data=df)
plt.title('Age vs. Total Spending')
plt.xlabel('Customer Age')
plt.ylabel('Total Spending')
plt.show()

# 4. Spending by Gender and Region (grouped bar chart)
plt.figure(figsize=(10,6))
sns.barplot(x='Gender', y='Total_Spending', hue='Region', data=df, estimator=sum)
plt.title('Total Spending by Gender and Region')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.show()
```

```python
#6. Export a final summary report in csv format.

df.to_csv("Cleaned_Customer_Purchases.csv", index=False) 
```
